```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('flight_data.csv')
```

```python
df.head()
```

```python
airline = df['airline'].tolist()
for i in range(len(airline)):
    # Only get the first word of the airline name
    airline[i] = airline[i].split()[0]
airline

df['airline'] = airline

src = df['src'].tolist()
for i in range(len(src)):
    src[i] = src[i].split()[-1]
src

df['src'] = src

dest = df['dest'].tolist()
for i in range(len(dest)):
    dest[i] = dest[i].split()[-1]
dest

df['dest'] = dest
```

```python
df['fare_class'].unique()
```

```python
price = df['price'].tolist()
for i in range(len(price)):
    price[i] = ''.join(filter(str.isdigit, price[i]))
    price[i] = int(price[i])

df['price'] = price
```

```python
fare_class = df['fare_class'].tolist()   

economy = ['eco', 'dlx']

for i in range(len(fare_class)):
    if "eco" in fare_class[i].lower() or "dlx" in fare_class[i].lower():
        fare_class[i] = "economy"
    else:
        fare_class[i] = "business"

fare_class  
df['fare_class'] = fare_class
```

```python
depart_time = df['depart_time'].tolist()
for i in range(len(depart_time)):
    depart_time[i] = depart_time[i].split(":")[0]
    if 6 <= int(depart_time[i]) < 12:
        depart_time[i] = "morning"
    elif 12 <= int(depart_time[i]) < 17:
        depart_time[i] = "afternoon"
    elif 17 <= int(depart_time[i]) < 22:
        depart_time[i] = "evening"
    else:
        depart_time[i] = "night"

depart_time
df['depart_time'] = depart_time
```

```python
arrival_time = df['arrival_time'].tolist()
for i in range(len(arrival_time)):
    arrival_time[i] = arrival_time[i].split(":")[0]
    if 6 <= int(arrival_time[i]) < 12:
        arrival_time[i] = "morning"
    elif 12 <= int(arrival_time[i]) < 17:
        arrival_time[i] = "afternoon"
    elif 17 <= int(arrival_time[i]) < 22:
        arrival_time[i] = "evening"
    else:
        arrival_time[i] = "night"

arrival_time
df['arrival_time'] = arrival_time
```

```python
for rows in range(0,len(df)):
    df.loc[rows, "depart_date"] = df.loc[rows, "depart_date"]+'-2024'
    df.loc[rows, "depart_date"] = df.loc[rows, "depart_date"].replace(' ','')
    df.loc[rows, "depart_date"] = df.loc[rows, "depart_date"].replace('th','-')
    df.loc[rows, "arrival_date"] = df.loc[rows, "arrival_date"]+'-2024'
    df.loc[rows, "arrival_date"] = df.loc[rows, "arrival_date"].replace(' ','')
    df.loc[rows, "arrival_date"] = df.loc[rows, "arrival_date"].replace('th','-')
```

```python
d_date = []
a_date = []
for date in pd.to_datetime(df['depart_date'], format='%d-%m-%Y').dt.date:
    if len(pd.bdate_range(date,date).strftime('%d-%m-%Y')) == 0:
        d_date.append('weekend')
    else:
        d_date.append('week_day')
for date in pd.to_datetime(df['depart_date'], format='%d-%m-%Y').dt.date:
    if len(pd.bdate_range(date,date).strftime('%d-%m-%Y')) == 0:
        a_date.append('weekend')
    else:
        a_date.append('week_day')
df['d_date'] = pd.DataFrame(d_date)
df['a_date'] = pd.DataFrame(a_date)
        
```
