```python
src = ["sgn", "han", "dad", "pqc", "cxr"]
departdate = "2024-05-29"

for s in src:
    dst = [d for d in src if d != s]
    for d in dst:
        try:
            path = r"D:\misc\chromedriver-win64\chromedriver.exe"
            content_XPATH = "//app-flight-list-detail/div/div[2]"
            price_XPATH = '//div[1]/div[2]/div[1]/div[5]/h4/b'
            XPATH = "//app-flight-list-detail/div/div/div/div/label[1]"

            url = f"https://www.ivivu.com/ve-may-bay/tim-kiem/ho-chi-minh-di-phu-quoc?from={s}&to={d}&departdate={departdate}&adult=1&child=0&infant=0&timeDepart=09:00&timeReturn=15:00"
            driver = webdriver.Chrome(path)
            driver.implicitly_wait(10)

            driver.get(url)
            driver.maximize_window()

            flight = driver.find_elements_by_xpath(XPATH)

            # wait = WebDriverWait(driver, 10)
            # wait.until(lambda d: flight.is_displayed())

            for i in flight:
                driver.execute_script("arguments[0].click();", i)
            price = driver.find_elements_by_xpath(price_XPATH)
            price_list = [p.text for p in price]
            content = driver.find_elements_by_xpath(content_XPATH)
            content_list = [c.text for c in content]

            flight_data = pd.DataFrame(content_list, columns=['content'])
            flight_data['price'] = price_list            
            flight_data.to_csv(f"flight_data_{s}_{d}.csv", index=False)
            print(f"flight_data_{s}_{d}.csv saved")
            driver.quit()
        except:
            print(f"Error at {s} to {d}")
            driver.quit()
        
```
